# Normal Distribution

```python
# normal distribution curve
mean = df['Current'].mean()
std = df['Current'].std(ddof=1)

x_min = df['Current'].min()
x_max = df['Current'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.plot(x, p,
         color="#E25CA8",
         linewidth=2)

plt.title('Normal Distribution of Current Values',
          font='Palatino Linotype',
          fontsize=12)
plt.xlabel('Current (mA)')
plt.ylabel('Probability Density')
plt.show
```

```python
# histogram plot
plt.hist(df['Current'], bins=5,
         density='true',
         alpha = 0.8,
         color= "#B76FA1",
         edgecolor= "#D4529C")

plt.title('Normal Distribution of Current Values',
          font='Palatino Linotype',
          fontsize=12)
plt.xlabel('Current (mA)')
plt.ylabel('Probability Density')

plt.plot(x, p,
         color="#BD00A7",
         linewidth=2)

plt.show
```

```python
# Standard normal distribution N(0,1)
# x-axis

zmean = df['Z-score'].mean()
zstd = df['Z-score'].std(ddof=1)

z_min = df['Z-score'].min()
z_max = df['Z-score'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, zmean, zstd)

plt.plot(x, p,
         color="#BD00A7",
         linewidth=2)

plt.title('Normal Distribution of Current Values N(0,1)',
          font='Courier New',
          fontsize=12)
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.show()
```
